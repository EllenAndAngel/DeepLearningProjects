```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
D = 2
```

```python
X = np.random.randn(N,D)
```

```python
X.shape
```

```python
ones = np.ones((N,1))
```

```python
ones.shape
```

```python
Xb = np.concatenate((ones,X),axis=1)
```

```python
Xb.shape
```

```python
w = np.random.randn(D + 1)
```

```python
w
```

```python
z = Xb.dot(w)
```

```python
len(z)
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
sig = sigmoid(z)
sig.
```

```python
plt.figure()
```

```python
plt.plot(sig)
plt.show()
```
