```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
image = cv2.imread('images/angel_standard.jpg')
```

```python
me = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('', me)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(me)
```

```python
row, col = np.shape(me)
```

```python
row
```

```python
col
```

```python
me.shape
```

```python
thresh = 250
```

```python
for i in range(0,row):
    for j in range (0, col):
        if me[i][j]>=thresh:
            me[i][j]=255
        else:
            me[i][j]=0
```
