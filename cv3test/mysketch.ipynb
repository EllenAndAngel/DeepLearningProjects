```python
import cv2
import numpy as np
```

```python
image = cv2.imread('images/IMG_4376.jpg')
```

```python
image.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(image, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
plt.imshow(image,cmap='gray_r')
```

```python
me_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(me_rgb)
```

```python
me_rgb_gray = cv2.cvtColor(me_rgb,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(me_rgb_gray, cmap='gray')
```

```python
# From original image to gray
me_bgr_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(me_bgr_gray, cmap='gray')
```

```python
me_gray = 255 - me_bgr_gray
```

```python
plt.imshow(me_gray, cmap='gray')
```

```python
me_gray_blur = cv2.GaussianBlur(me_gray,(21,21),0,0)
```

```python
plt.imshow(me_gray_blur, cmap='gray')
```

```python
me_canny_edges =cv2.Canny(me_gray_blur,10,70)
```

```python
plt.imshow(me_canny_edges, cmap='gray_r')
```

```python
_,mask = cv2.threshold(me_canny_edges,70,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(mask, cmap='gray')
```
