```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
fname = 'images/angel_standard.jpg'
image = Image.open(fname).convert("L")
arr = np.asarray(image)
plt.imshow(arr, cmap='gray_r')
plt.show()
```

```python
row, col = np.shape(image)
print(row, col)
```

```python
arr
```

```python
import cv2
```

```python
small = cv2.pyrDown(arr)
small.shape
```

```python
plt.imshow(small,cmap='gray_r')
plt.show()
```

```python
me_gray_blur = cv2.bilateralFilter(small,9,75,75)
```

```python
plt.imshow(me_gray_blur,cmap='gray')
plt.show()
```

```python
me_canny_edges =cv2.Canny(me_gray_blur,10,70)
```

```python
plt.imshow(me_canny_edges,cmap='gray_r')
plt.show()
```

```python
_,mask = cv2.threshold(me_canny_edges,70,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(mask,cmap='gray_r')
plt.show()
```
